```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Stability Diagram
def a0 (q):
    p02 = -1.0/2.0
    p04 = 7.0/128.0
    p06 = -29.0/2304.0
    p08 = 68687.0/18874368.0
    return p02*q**2 + p04*q**4 + p06*q**6 + p08*q**8

def a1 (q):
    p10 = 1.0
    p11 = -1.0
    p12 = -1.0/8.0
    p13 = 1.0/64.0
    p14 = -1.0/1536.0
    p15 = -11.0/35864.0
    return p10 + p11*q + p12*q**2 + p13*q**3 + p14*q**4 + p15*q**5
```

```python
q = np.array(range(10,140))/100.

plt.plot(q,a0(q),label='a_0(q)')
plt.plot(q,a1(q),label='a_1(q)')
plt.plot(q,-2.0*a1(q/2.0),label='-2*a_1(q/2)')
plt.plot(q,-2.0*a0(-q/2.0),label='-2*a_0(-q/2)')

plt.plot(0.53,-0.01,'o',label='Ion')

plt.title('Stability Diagram')
plt.xlabel('q')
plt.ylabel('a')

plt.grid(alpha=.4,linestyle='--')
plt.legend()
plt.show()
```
